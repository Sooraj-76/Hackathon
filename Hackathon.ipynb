```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# Load the dataset
data = pd.read_csv('hacktrain.csv')
data
```

```python
data.isnull().sum()
```

```python
data = data.drop_duplicates()
```

```python
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns
imputer = SimpleImputer(strategy='median')
data[numeric_cols] = imputer.fit_transform(data[numeric_cols])
```

```python
print(data.isnull().sum())
```

```python
# Encode class column (if it's categorical)
label_encoder = LabelEncoder()
data['class'] = label_encoder.fit_transform(data['class'])
```

```python
# Separate features and target variable
X = data.drop(columns=['class'])
y = data['class']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
# Normalize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize and train the model
model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=200)
model.fit(X_train_scaled, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test_scaled)
```

```python
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
print(classification_report(
    y_test,
    y_pred,
    labels=list(range(len(label_encoder.classes_))),
    target_names=label_encoder.classes_
))
```

```python
test_data = pd.read_csv('hacktest.csv')
test_data
```

```python
# Drop the 'ID' column and store the ID for later use
ID = test_data['ID']
test_data.drop(columns=['ID'], inplace=True) # Using `columns` parameter for clarity
```

```python
# **Model Prediction**
# Ensure test_data columns match the training features
test_data_aligned = test_data.reindex(columns=X_train.columns, fill_value=0)
# Apply the same scaler as used for training
test_data_scaled = scaler.transform(test_data_aligned)
# Predict using the trained model
y_pred = model.predict(test_data_scaled)
# Decode the predicted labels to original class names
y_decoded = label_encoder.inverse_transform(y_pred)
```

```python
# **Create Result DataFrame**
result = pd.DataFrame({
    'ID': ID,
    'class': y_decoded
})
result
```

```python
# **Output the Results**
result.to_csv("submission.csv", index=False) 
```
